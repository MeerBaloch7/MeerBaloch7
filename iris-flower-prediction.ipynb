# 1: import Lbraries

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
```

# 2: Loading of data

```python
data= pd.read_csv('/kaggle/input/iris-flower-dataset/IRIS.csv')
data.head()
```

# 3: Overview of Data

```python
data.describe()
```

```python
data.columns
```

```python
# checking for null values
missing=data.isnull().sum()
percentage_missing=missing/len(data)*100
print(20*"-" + "Missing Values".center(20) + 20*"-","\n",missing)
print(20*"-" + "Percentage Missing Values".center(20) + 20*"-","\n",percentage_missing)

#print("Missing values:\n",missing,'\n--------\n',"Percentage:\n",percentage_missing)
```

```python
data['species'] # target
```

```python
data['species']=data['species'].str.replace('Iris-','')
data['species']
```

### Transforming of species into numericals

```python
label=LabelEncoder()
data.species=label.fit_transform(data.species)
data.species
```

# 3: Data Visualization

```python
sns.pairplot(data=data,hue='species')
```

```python
data.hist()
```

```python
corr=data.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("Iris Flower Prediction")
```

# 5: Building Model

```python
# features
X=data.drop(["species"],axis=1)
X
```

```python
# target
Y=data.species
Y
```

```python
# spliting of data
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=41)
```

### 5.1 Logistic regression

```python
Lg= LogisticRegression()
Lg.fit(X_train,Y_train)
predict=Lg.predict(X_test)
Lg_Score=accuracy_score(Y_test,predict)
print(" accuracy score of logistic Regression: ",Lg_Score)
```

### 5.2: Decision Tree Classifier

```python
DTC=DecisionTreeClassifier()
DTC.fit(X_train,Y_train)
pred= DTC.predict(X_test)
DTC_Score= accuracy_score(Y_test,pred)
print(" accuracy score of Decision tree Classifier: ",DTC_Score)
```

### 5.3: Random Forest Classifier

```python
RFC=RandomForestClassifier()
RFC.fit(X_train,Y_train)
predicts=RFC.predict(X_test)
RFC_Score= accuracy_score(Y_test,predicts)
print(" accuracy score of Random Forest Classifier: ",RFC_Score)
```

# 6: Scores

```python
print("Logistic Regression: ",Lg_Score)
print("Random Forest Classifier: ",RFC_Score)
print(" Decision Tree Classifier: ",DTC_Score)
```
